```python
from googletrans import Translator
```

```python
translator = Translator()
```

```python
import pandas as pd
```

```python
df_product_br = pd.read_csv('olist_products_dataset_br.csv', sep=',')
df_product_br.head()
```

```python
df_product_en = df_product_br.copy()
```

```python
df_product_en.head()
```

```python
df_product_en.rename(columns=lambda x:translator.translate(x).text, inplace=True)

df_product_en.columns
```

```python
df_product_en['transalted'] = df_product_en['product_category_name'].apply(lambda x: translator.translate(x, dest='en').text)
```

```python
df_product_en.head()
```

```python
df_product_en.to_csv('olist_products_dataset_en.csv')
```
